# Seaborn

**Seaborn** is a Python data visualization library built on top of Matplotlib. It provides a higher-level interface for drawing attractive and informative statistical graphics with less code.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**`sns.color_palette()`** returns a list of colors. Seaborn has many built-in palettes like `deep`, `muted`, `pastel`, `bright`, `dark`, `colorblind`.


```python
sns.color_palette('deep')
sns.color_palette('muted')
sns.color_palette('pastel')
sns.color_palette('bright')
sns.color_palette('dark')
sns.color_palette('colorblind')
```

**`sns.get_dataset_names()`** returns a list of all built-in datasets available in Seaborn that can be loaded directly without any external file.


```python
sns.get_dataset_names()
```

# Load Dataset **tips**

**`sns.load_dataset()`** loads a built-in Seaborn dataset directly into a DataFrame without needing any external file.


```python
tips = sns.load_dataset('tips')
tips
```


**`head()`** returns the first 5 rows of the DataFrame to get a quick preview of the data.


```python
tips.head(5)
```

**Scatter Plot**

**`sns.scatterplot()`** creates a scatter plot. `data` specifies the DataFrame, `x` and `y` specify the columns for each axis.


```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

**`hue`** separates data points by a categorical column and assigns different colors to each category.


```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

**`size`** parameter scales the size of each data point based on a numerical column's value.


```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size='size')
plt.show()
```

**`style`** changes the marker shape for each category in the specified column.


```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='smoker')
plt.show()
```

**`sns.relplot()`** is a figure-level function for visualizing statistical relationships. **`kind='scatter'`** creates a scatter plot with more customization options than `sns.scatterplot()`.


```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='smoker', kind='scatter')
plt.show()
```

**`col`** splits the plot into multiple subplots based on a categorical column — one subplot for each unique value in that column.


```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='smoker', kind='scatter', col='day')
plt.show()
```

**`col_wrap`** controls how many plots appear in each row before wrapping to the next row.


```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='smoker', kind='scatter', col='day', col_wrap=2)
plt.show()
```

# Load Dataset **dowjons**

**`dowjones`** is a built-in Seaborn dataset that contains the Dow Jones Industrial Average stock prices over time. It has two columns — `Date` and `Price`.

```python
dowjones = sns.load_dataset('dowjones')
dowjones
```

```python
dowjones.head(10)
```

**Line Plot**

**`sns.lineplot()`** creates a line graph to show trends over time. Best used for time series data like stock prices.


```python
sns.lineplot(data=dowjones, x='Date', y='Price')
plt.show()
```

**`sns.relplot()`** with **`kind='line'`** and **`markers=True`** shows the line plot with dots at each data point.


```python
sns.relplot(data=dowjones, x='Date', y='Price', kind='line', markers=True)
plt.show()
```

**Displot**

**`sns.displot()`** shows the distribution of a single variable. It plots how frequently each value appears in the data.


```python
sns.displot(data=tips, x='tip')
plt.show()
```

**`col='sex'`** splits the distribution plot into separate panels for each category in the `sex` column — one for Male and one for Female, making it easy to compare distributions side by side.

```python
sns.displot(data=tips, x='tip', col='sex')
plt.show()
```

**`kind='hist'`** explicitly draws a histogram showing the frequency of values in each bin.


```python
sns.displot(data=tips, x='tip', col='sex', kind='hist')
plt.show()
```

**Hisplot**

**`sns.histplot()`** is a axes-level function that draws a histogram directly without needing `kind` parameter.


```python
sns.histplot(data=tips, x='tip')
plt.show()
```

**`hue='sex'`** separates the histogram bars by category and assigns different colors to each, so Male and Female distributions are visible on the same plot simultaneously.

```python
sns.histplot(data=tips, x='tip', hue='sex')
plt.show()
```

**Countplot**

**`sns.countplot()`** counts the occurrences of each category and displays them as bars. It is the most common categorical plot.


```python
sns.countplot(data=tips, x='day')
plt.show()
```

**Stripplot**

**`sns.stripplot()`** shows all individual data points for each category. It is useful to see the actual distribution of data within each category.


```python
sns.stripplot(data=tips, x='day', y='tip')
plt.show()
```

**Box Plot**

**`sns.boxplot()`** shows the distribution of data through quartiles. It displays the median, spread and outliers of the data for each category.


```python
sns.boxplot(data=tips, x='tip', y='day')
plt.show()
```

**`hue='sex'`** splits each box into two — one for Male and one for Female — placed side by side for each day, making it easy to compare tip distributions between genders across different days.

```python
sns.boxplot(data=tips, x='tip', y='day', hue='sex')
plt.show()
```

**`fill=False`** removes the color fill from the boxes and shows only the outline, making the plot cleaner.


```python
sns.boxplot(data=tips, x='tip', y='day', hue='sex', fill=False)
plt.show()
```

**Violin Plot**

**`sns.violinplot()`** combines a box plot and a kernel density plot. It shows the distribution shape of the data for each category — wider sections mean more data points at that value.


```python
sns.violinplot(data=tips, x='tip', y='day', hue='sex',fill=False)
plt.show()
```

**Bar Plot**

**`sns.barplot()`** shows the mean value of a numeric column for each category with a confidence interval bar showing the uncertainty around the mean.


```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.show()
```

**`sns.countplot()`** counts the number of occurrences of each category in the `day` column. **`hue='sex'`** splits each bar into Male and Female so we can compare how many people of each gender visited on each day.

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

**Pair Plot**

**`sns.pairplot()`** plots relationships between all numeric columns in the dataset at once. Diagonal shows the distribution of each column and off-diagonal shows scatter plots between each pair of columns.


```python
sns.pairplot(tips)
plt.show()
```

**`hue='sex'`** colors all scatter plots and distribution curves by gender, making it easy to compare patterns between Male and Female across all variable combinations at once.

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
import numpy as np
```

**Creating a 4x4 array with large diagonal values** so the heatmap shows a clear pattern along the diagonal.


```python
arr = np.random.randint(1, 10, (4, 4))
np.fill_diagonal(arr, 100)
arr
```

```python
import matplotlib.pyplot as plt
```

**HeatMap**

**`sns.heatmap()`** visualizes a 2D array as a color-coded grid. **`annot=True`** displays the actual values inside each cell.


```python
plt.figure(figsize=(6, 4))
sns.heatmap(arr)
plt.show()
```

**`fmt='d'`** formats the annotation as integer. **`cmap`** sets the color scheme. **`cbar=True`** displays the color bar on the side showing the value scale.


```python
plt.figure()
sns.heatmap(arr, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.show()
```

**`xticklabels`** and `yticklabels`** set the labels for x and y axis ticks. **`plt.xlabel()`** and **`plt.ylabel()`** add axis titles.


```python
labels = ['Dog', 'Cow', 'Rat', 'Rabbit']

plt.figure()
sns.heatmap(arr, annot=True, fmt='d', cmap='Blues', cbar=True, xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

**Correlation Matrix**

**`df.corr()`** calculates the correlation between all numeric columns. Values range from -1 to 1 — closer to 1 means strong positive correlation, closer to -1 means strong negative correlation.


```python
plt.figure()
sns.heatmap(tips.corr(numeric_only=True), annot=True, fmt='.2f', cmap='Blues', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

**Linear Regression**

**`sns.regplot()`** plots a scatter plot with a linear regression line fitted to the data, showing the trend between two variables.


```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```
