# **Matplotlib**

**Matplotlib** is a Python library used for creating static, animated and interactive visualizations. **`pyplot`** is the most commonly used module in Matplotlib, imported as `plt`.



```python
import matplotlib.pyplot as plt
```

**Import NumPy**

```python
import numpy as np
```

# Single Graph

**`range(1, 9)`** generates numbers from 1 to 8. **`list()`** converts it into a list which can be used as x-axis values for plotting.

```python
x = list(range(1, 9))
x
```

**`np.random.randint()`** generates random integers between a given range. **`tolist()`** converts the NumPy array into a plain Python list.


```python
y = np.random.randint(52, 70, size=8).tolist()
y
```

**`plt.plot()`** creates a line graph by connecting data points with a line. `x` is the horizontal axis and `y` is the vertical axis. **`plt.show()`** displays the final plot.

```python
plt.plot(x, y)
plt.show()
```

**`plt.xlabel()`** and **`plt.ylabel()`** add labels to the x and y axis respectively to make the plot more readable.
**`plt.title()`** adds a title to the top of the plot.


```python
plt.plot(x, y)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

**`color`** parameter changes the color of the line in the plot.


```python
plt.plot(x, y, color='red')
```

**`linestyle`** changes the style of the line in the plot. `--` creates a dashed line.


```python
plt.plot(x, y, color='red', linestyle='--')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

**`marker`** adds a symbol at each data point. `'s'` represents a square marker.

```python
plt.plot(x, y, color='red', linestyle='--', marker='s')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

**`markersize`** controls the size of the marker at each data point.



```python
plt.plot(x, y, color='red', linestyle='--', marker='s', markersize=10)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

**`'r--s'`** is a shorthand format in Matplotlib — `r` means red color, `--` means dashed line and `s` means square marker. All three styles are combined in a single string instead of writing separate parameters.

```python
plt.plot(x, y, 'r--s', markersize=10)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

**`alpha`** controls the transparency of the line and markers. Value ranges from 0 (fully transparent) to 1 (fully opaque).


```python
plt.plot(x, y, 'r--s', markersize=10, alpha=0.5)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.show()
```

# Multiple Graph

**`list(range(1,9))`** creates x axis values from 1 to 8. **`np.random.randint()`** generates random temperatures and **`.tolist()`** converts them to Python lists. **`avg_temp`** is calculated by converting min and max to NumPy arrays so element wise addition and division is possible, giving the average temperature for each day.

```python
x = list(range(1, 9))
min_temp = np.random.randint(10, 30, size=8).tolist()
max_temp = np.random.randint(50, 70, size=8).tolist()
avg_temp = (np.array(max_temp) + np.array(min_temp)) / 2
```

**Multiple plots** can be drawn on the same graph by calling `plt.plot()` multiple times. **`label`** gives each line a name and **`plt.legend()`** displays those labels as a legend on the graph so each line can be identified easily.

```python
plt.plot(x, min_temp, 'b--s', markersize=8, label='Minimum')
plt.plot(x, max_temp, 'r--s', markersize=8, label='Maximum')
plt.plot(x, avg_temp, 'g--s', markersize=8, label='Average')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.legend()
plt.show()
```

**`shadow=True`** adds a shadow effect behind the legend box making it more visible.


```python
plt.plot(x, min_temp, 'b--s', markersize=8, label='Minimum')
plt.plot(x, max_temp, 'r--s', markersize=8, label='Maximum')
plt.plot(x, avg_temp, 'g--s', markersize=8, label='Average')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature over Days')
plt.legend(shadow=True)
plt.show()
```

# Bar Chart

**`plt.bar()`** creates a vertical bar chart. Each bar represents a category and its height represents the value.


```python
companies = ['Microsoft', 'Facebook', 'Google', 'Tesla']
revenue = [198, 117, 257, 81]

plt.bar(companies, revenue, color='blue')
plt.xlabel('Companies')
plt.ylabel('Revenue (Billion $)')
plt.title('Company Revenue')
plt.show()
```

**`color`** parameter accepts a list of colors when multiple bars are present — each color in the list is applied to the corresponding bar in order.

```python
companies = ['Microsoft', 'Facebook', 'Google', 'Tesla']
revenue = [198, 117, 257, 81]

plt.bar(companies, revenue, color=['blue', 'red', 'green', 'orange'])
plt.xlabel('Companies')
plt.ylabel('Revenue (Billion $)')
plt.title('Company Revenue')
plt.show()
```

**`label`** gives each bar a name and **`plt.legend()`** displays them. Since all bars have different colors, each color needs a separate `plt.bar()` call with its own label.


```python
plt.bar('Microsoft', 198, color='blue', label='Microsoft')
plt.bar('Facebook', 117, color='red', label='Facebook')
plt.bar('Google', 257, color='green', label='Google')
plt.bar('Tesla', 81, color='orange', label='Tesla')
plt.xlabel('Companies')
plt.ylabel('Revenue (Billion $)')
plt.title('Company Revenue')
plt.legend()
plt.show()
```

**`plt.text(i, revenue[i] + 2, str(revenue[i]))`** places text on top of each bar. `i` is the x position, `revenue[i] + 2` places the text slightly above the bar, `str(revenue[i])` converts the number to string for display. **`ha='center'`** centers the text horizontally over each bar. **`range(len(companies))`** loops through each bar one by one.

```python
companies = ['Microsoft', 'Facebook', 'Google', 'Tesla']
revenue = [198, 117, 257, 81]

bars = plt.bar(companies, revenue, color=['blue', 'red', 'green', 'orange'])

for i in range(len(companies)):
    plt.text(i, revenue[i] + 2, str(revenue[i]), ha='center')

plt.xlabel('Companies')
plt.ylabel('Revenue (Billion $)')
plt.title('Company Revenue')
plt.show()
```

**Profit and Revenue are in the same bar**

**`plt.bar()`** with a single x position stacks both revenue and profit as separate bars for each company. **`bottom`** parameter places profit bar on top of revenue bar for stacked visualization.


```python
companies = ['Microsoft', 'Facebook', 'Google', 'Tesla']
revenue = [198, 117, 257, 81]
profit = [72, 39, 76, 12]
colors = ['blue', 'red', 'green', 'orange']

x = np.arange(len(companies))

for i in range(len(companies)):
    plt.bar(x[i], revenue[i], color=colors[i], label=companies[i])
    plt.bar(x[i], profit[i], color='yellow', bottom=revenue[i])

for i in range(len(companies)):
    plt.text(i, revenue[i] / 2, str(revenue[i]), ha='center', color='white')
    plt.text(i, revenue[i] + profit[i] + 2, str(profit[i]), ha='center')

plt.xticks(x, companies)
plt.xlabel('Companies')
plt.ylabel('Billion $')
plt.title('Company Revenue vs Profit')
plt.legend()
plt.show()
```

**Profit and Revenue are in the different bar**

**`np.arange(len(companies))`** creates numeric positions for each company on x axis. **`x - 0.2`** shifts the revenue bar slightly to the left and **`x + 0.2`** shifts the profit bar slightly to the right so both bars sit side by side for each company. **`width=0.4`** controls the width of each bar. **`plt.xticks(x, companies)`** replaces the numeric positions with company names on the x axis.

```python
companies = ['Microsoft', 'Facebook', 'Google', 'Tesla']
revenue = [198, 117, 257, 81]
profit = [72, 39, 76, 12]

x = np.arange(len(companies))

plt.bar(x - 0.2, revenue, width=0.4, label='Revenue')
plt.bar(x + 0.2, profit, width=0.4, label='Profit')

for i in range(len(companies)):
    plt.text(i - 0.2, revenue[i] + 2, str(revenue[i]), ha='center')
    plt.text(i + 0.2, profit[i] + 2, str(profit[i]), ha='center')

plt.xticks(x, companies)
plt.xlabel('Companies')
plt.ylabel('Billion $')
plt.title('Company Revenue vs Profit')
plt.legend()
plt.show()
```

**Horizontal Bar Chart**

**`plt.barh()`** creates a horizontal bar chart. Everything is same as `plt.bar()` but x and y axis are swapped.


```python
plt.barh(companies, revenue, color=['blue', 'red', 'green', 'orange'])
plt.xlabel('Revenue (Billion $)')
plt.ylabel('Companies')
plt.title('Company Revenue')
plt.show()
```

# Histogram

**Step 1 — Data generate**

**np.random.randn()** generates random data following a normal distribution — most values cluster around the mean.

```python
data = np.random.randn(1000)
```

**Step 2 — Basic histogram**

**plt.hist()**   divides data into bins and counts how many values fall in each bin.

```python
plt.hist(data, bins=30)
plt.show()
```

**Step 3 — Color + title + label**

```python
plt.hist(data, bins=30, color='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

**Step 4 — Edge color add**

**edgecolor** adds a border around each bin making it easier to distinguish.

```python
plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

**Step 5 — Transparency add**

**alpha** makes the bars slightly transparent which is useful when multiple histograms overlap.

```python
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

# Pie Chart

**`plt.pie()`** creates a pie chart. **`labels`** adds category names, **`autopct='%1.1f%%'`** shows percentage on each slice.


```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Programming Language Popularity')
plt.show()
```

**`pctdistance`** controls where the percentage text appears — values less than 1 place it inside the slice. **`labeldistance`** controls where the label appears — values greater than 1 place it outside the slice.



```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=1.2, labeldistance=0.5)
plt.title('Programming Language Popularity')
plt.show()
```

**`explode`** pulls out a specific slice from the pie chart to highlight or focus on it. The value represents how far the slice is pulled out.


```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
explode = (0.1, 0, 0, 0)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)
plt.title('Programming Language Popularity')
plt.show()
```

# Scatter Plot

**Step 1 — Data generate**

**`np.random.rand()`** generates random float values between 0 and 1 for both x and y axis.


```python
x = np.random.rand(50)
y = np.random.rand(50)
```

**Step 2 — Basic scatter plot**

**`plt.scatter()`** plots individual data points without connecting them.


```python
plt.scatter(x, y)
plt.show()
```



**Step 3 — Color + title + label**


```python
plt.scatter(x, y, color='red')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.show()
```


**Step 4 — Marker size change**

**`s`** controls the size of each data point in the scatter plot.


```python
plt.scatter(x, y, color='red', s=100)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.show()
```



**Step 5 — Transparency add**



```python
plt.scatter(x, y, color='red', s=100, alpha=0.5)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.show()
```
