# **Pandas**

## Pandas Data Structures

Pandas has **2 core data structures:**

---

### 1. Series
A **one-dimensional** labeled array. Think of it like a single column in Excel.

- Has an **index** (labels) + **values**
- Can hold any data type (int, float, string, etc.)

---

### 2. DataFrame
A **two-dimensional** labeled table. Think of it like a full Excel sheet.

- Has **rows** and **columns**
- Each column is actually a Series
- Most commonly used in data analysis

---

### Quick Comparison

| | Series | DataFrame |
|---|---|---|
| Dimension | 1D | 2D |
| Like | One column | Full table |
| Has index | ✅ | ✅ (rows + columns) |



# **Creating a Series (Without Index)**



```python
s1 = pd.Series([10, 20, 30, 40])
s1
```

# **Creating a Series (With Index)**

```python
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s1
```

A Series is a 1D labeled array in Pandas. We
 can create it directly from a Python list using **pd.Series()**.


```python
my_list = [10, 20, 30, 40]
s = pd.Series(my_list)
s
```

# **Creating a DataFrame**

A DataFrame is a 2D labeled table with rows and columns. Each column is a Series, and it works like an Excel spreadsheet.

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85, 90, 78]
}

df = pd.DataFrame(data)
df
```
