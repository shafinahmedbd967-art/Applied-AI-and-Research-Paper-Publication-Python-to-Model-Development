**`pd.DataFrame()`** creates a DataFrame manually. `None` is used to represent missing/null values in Python which Pandas reads as `NaN`.


```python
data = {
    'Name': ['Alice', 'Bob', None, 'David', None],
    'Age': [25, None, 35, None, 28],
    'Score': [85, 90, None, 78, None]
}

df = pd.DataFrame(data)
df
```

**`ffill()`** (forward fill) replaces null values with the **previous row's value**. It fills forward, meaning it takes the last valid value above the null and fills it down.


```python
ffill_df = df.ffill()
ffill_df
```

**`bfill()`** (backward fill) replaces null values with the **next row's value**. It fills backward, meaning it takes the next valid value below the null and fills it up.

```python
bfill_df = df.bfill()
bfill_df
```

**`describe()`** helps detect outliers by comparing **min, max** with **mean and std**. If max is much higher or min is much lower than mean, there are likely outliers.

**Step 2** — Outlier Detect using IQR method

**IQR (Interquartile Range)** = Q3 - Q1. Any value below `Q1 - 1.5*IQR` or above `Q3 + 1.5*IQR` is considered an outlier.

```python
df.describe()
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Age'] < Q1 - 1.5 * IQR) | (df['Age'] > Q3 + 1.5 * IQR)]
outliers
```
